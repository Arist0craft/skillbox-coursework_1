#### Объединение с последним датасетом
Для подсчета некоторых метрик потребуется объединить основной датасет с `df_students`

```python
# Merge datasets
field = ["title", "student_id"]
df_student_age = \
    main[field]\
    .groupby(field)\
    .count()\
    .reset_index()\
    .merge(df_students, left_on="student_id", right_on="id")

df_student_age
```

Добавим столбец возраст, исходя из сегодняшней даты, но вначале проверим на отсутствующие даты дней рождения

```python
# Check if birthday has NaN values
df_student_age[df_student_age.birthday.isna()]
```

```python
# Let's trim them
student_age_trimmed = df_student_age.dropna(subset=["birthday"])

# pd.to_datetime won't work on field birthday, because some of them out of
# pandas timedelta range. Let's apply function which calculating age
def get_age(x):
    birthday = dt.strptime(x, "%Y-%m-%d")
    today = dt.today()
    return (today.year - birthday.year - 
            ((today.month, today.day) < (birthday.month, birthday.day)))


student_age_trimmed["age"] = student_age_trimmed["birthday"].map(get_age)
pd.DataFrame(student_age_trimmed.age)
```

Теперь перейдем к подсчету метрик

- Минимальный, максимальный, средний, медианный возраст студентов

```python
student_age_trimmed.age.describe()
```

- Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

```python
student_age_trimmed.groupby("title")["age"]\
    .agg(min="min", max="max", mean="mean", median="median")
```

### 1.3 Графический анализ.
Теперь построим графическое отображение наших данных

```python
plot_data =\
    student_age_trimmed\
        .groupby("title")["student_id"]\
        .count().reset_index()

colors = cm.rainbow(np.linspace(0, 1, plot_data.shape[0]))

fig = plt.figure(figsize=(13,12))
fig.set_facecolor("white")
plt.title("Количество студентов на каждом курсе")

plt.bar(
    data=plot_data,
    x="title",
    height="student_id",
    color=colors,
    figure=fig
)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Количество студентов")
plt.show()
```

```python
plot_data = plot_data.sort_values("student_id")

fig, ax = plt.subplots(figsize=(13, 12))

# Set figure background color to white for better visual with dark theme
fig.set_facecolor("white")
plt.title("Количество студентов на каждом курсе")

barh = ax.barh(
    data=plot_data,
    y="title",
    width="student_id",
    color=colors
)
ax.set_xlabel("Количество студентов")


# Median line
median = ax.axvline(x=plot_data.median()[0], color="gray")

# Remove frame from axis
for k in ax.spines:
    ax.spines[k].set_color(None)

# Legend
plt.legend(barh, plot_data.title.values)
plt.show()
```
