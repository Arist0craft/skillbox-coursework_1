#### Объединение с последним датасетом
Для подсчета некоторых метрик потребуется объединить основной датасет с `df_students`

```python
# Merge datasets
field = ["title", "student_id"]
df_student_age = \
    main[field]\
    .groupby(field)\
    .count()\
    .reset_index()\
    .merge(df_students, left_on="student_id", right_on="id")

df_student_age
```

Добавим столбец возраст, исходя из сегодняшней даты, но вначале проверим на отсутствующие даты дней рождения

```python
# Check if birthday has NaN values
df_student_age[df_student_age.birthday.isna()]
```

```python
# Let's trim them
student_age_trimmed = df_student_age.dropna(subset=["birthday"])

# pd.to_datetime won't work on field birthday, because some of them out of
# pandas timedelta range. Let's apply function which calculating age
def get_age(x):
    birthday = dt.strptime(x, "%Y-%m-%d")
    today = dt.today()
    return (today.year - birthday.year - 
            ((today.month, today.day) < (birthday.month, birthday.day)))


student_age_trimmed["age"] = student_age_trimmed["birthday"].map(get_age)
pd.DataFrame(student_age_trimmed.age)
```

```python
student_age_trimmed.groupby("title")["age"]\
    .agg(min="min", max="max", mean="mean", median="median")
```

### 1.3 Графический анализ.
Теперь построим графическое отображение наших данных

```python
plot_data =\
    student_age_trimmed\
        .groupby("title")["student_id"]\
        .count().reset_index()

colors = cm.rainbow(np.linspace(0, 1, plot_data.shape[0]))

fig = plt.figure(figsize=(13,12))
fig.set_facecolor("white")
plt.title("Количество студентов на каждом курсе")

plt.bar(
    data=plot_data,
    x="title",
    height="student_id",
    color=colors,
    figure=fig
)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Количество студентов")
plt.show()
```

### 2.1 Динамика количества новых студентов
Выберем законченные домашние задания. Далее сгруппируем по названию курса и `student_id` получим все домашние задания в рамках курса. Отсортируем и выберем самое первое. Дальше снова перегруппировываем по названию курса и месяцу года, считаем количество законченных первых домашних заданий студентов = количество новых студентов.

```python
### YOUR CODE HERE ###
def to_datetime(x):
    x["start_date"] = parser.parse(x["start_date"])
    x["finish_date"] = parser.parse(x["finish_date"])
    return x

def select_first(x):
    return x.dt.strftime("%Y-%m").sort_values().values[0]


done_homeworks = main[(main.status == "done") & (main.is_homework == True)]
done_homeworks = done_homeworks.apply(to_datetime, axis=1)
plot_data =\
    done_homeworks\
        .groupby(["title", "student_id"])["finish_date"]\
        .agg(select_first)\
        .reset_index()\
        .groupby(["title", "finish_date"])\
        .count()\
        .reset_index()
plot_data
```

***

### 2.2 Графики динамики количества новых студентов

```python
axes_count = len(plot_data.title.unique())

colors = cm.rainbow(np.linspace(0, 1, axes_count))

fig, axes = plt.subplots(
    axes_count, 1,
    sharex=False,
    figsize=(12, 22),
    constrained_layout=True
    )

fig.set_facecolor("white")
fig.suptitle("Динамика количества новых студентов")

for ax, (title, group), color in zip(axes, plot_data.groupby("title"), colors):
    group.plot("finish_date", "student_id", c=color, label=title, ax=ax)
    ax.set_xlabel(None)

plt.xlabel("Месяц года")
plt.ylabel("Количество новых студентов")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 12))
fig.set_facecolor("white")
plt.title("Динамика количества новых студентов")

for title, group in plot_data.groupby("title"):
    group.plot("finish_date", "student_id", label=title, ax=ax)

for k in ax.spines:
    ax.spines[k].set_color(None)

plt.xlabel("Месяц года")
plt.ylabel("Количество новых студентов")
plt.show()
```

```python
def to_date_range(x):
    return pd.date_range(
        x["start_date"].strftime("%Y-%m"),
        x["finish_date"].strftime("%Y-%m"),
        freq="MS"
        ).strftime("%Y-%m").to_list()
```

```python
plot_data = done_homeworks[["title", "progress_id", "start_date", "finish_date"]]
plot_data["work_month"] = plot_data.apply(to_date_range, axis=1)
plot_data =\
    plot_data\
        .explode("work_month")\
        .groupby(["title", "work_month"])["progress_id"]\
        .count()\
        .reset_index()

plot_data
```
